````python
#need to subset sb_inat into something that can be plotted

#lets say we wanted to know more about coastal birds in santa barbara
#found that birds are within the "Aves" class (highest ecological group that all birds share in common0

#see if Aves class is within this dataset
sb_classes = sb_inat['class']
sb_classes
````

````python
#isolate Aves observations from sb_classes data
spp_count = sb_inat.groupby("class")["species"]
spp_count_df = pd.DataFrame(spp_count)

#this gave the number of species within each class when i used.unique()
#produces a pandas.core.groupby.generic.SeriesGroupBy object at 0x7f58d9f70d00

#i want the number of birds. which is the 'aves' class

#gets the species within class of Aves out!
aves_spp = spp_count.get_group("Aves")  #can use this obj and join to lat and lon data in sb_inat
````

````python
#if i can isolate the df to just: bird spp name, lat, lon, i could try to overlay that?
aves_spp
#this gives all species of Aves class
````

````python
#select lat and lon from the columns 
lat_lon = sb_inat[["decimalLatitude", "decimalLongitude"]]
#this is all of the lat and lons in the dataset of sb_inat. 
#we want the ones only associated with the Aves! how can we do this. 

#join aves_spp and lat_lon
````

````python
#crop the initial data to just class, spp, and location information 
sb_inat_cropped = sb_inat[['class','species','decimalLatitude','decimalLongitude']]
sb_inat_cropped
````

````python
#now we want to isolate our table to only observations of class 'Aves'
sb_aves_locs = sb_inat_cropped[sb_inat_cropped['class'] == 'Aves']
sb_aves = gpd.GeoDataFrame(sb_aves_locs)
sb_aves_spp = sb_aves_locs[['species']]
sb_aves_spp
````

````python
#need to convert coordinates into actual coordinates, not just data points
coords = gpd.points_from_xy( sb_aves_locs['decimalLongitude'], sb_aves_locs['decimalLatitude'], crs = "EPSG:4326")


sb_aves_gdf = gpd.GeoDataFrame(sb_aves_spp, geometry=coords)
sb_aves_gdf
````

````python
sb_aves_gdf.plot(aspect = 1,
                color = "orchid")
````

````python
#setting up shp file path for CA state boundary (found online - https://data.ca.gov/dataset/ca-geographic-boundaries/resource/3db1e426-fb51-44f5-82d5-a54d7c6e188b)
shp_data = gpd.read_file("/courses/EDS214/EDS220_gbffs/ca-state-boundary/CA_State_TIGER2016.shp")

#produces an outline of CA 
shp_data.plot()
````

````python
shp_data.loc[:, 'geometry'].plot()
````

````python
shp_data
````

````python
#need to convert shp file into a lat/lon situation instead of meters
shp_reproj = shp_data.to_crs("epsg:4326") 
shp_reproj.plot()
````

````python
#check that this worked and that both data are on the same crs
shp_reproj.crs
````

````python
#make outline of sb to overlay inaturalist data onto

#defining size of output figure
fig, ax = plt.subplots(figsize=(14,6))

#setting boundaries
xmin = -120.8
xmax = -119.1
ymin = 34
ymax = 35

#feed min and max thru into limits for x and y axis
xlim = (xmin, xmax)
ylim = (ymin, ymax)

#set axis to x and y limits
ax.set_xlim(xlim)
ax.set_ylim(ylim)

#sb map
plt.title("iNaturalist Research-Grade Observations of Birds in Santa Barbara")
shp_reproj.plot(legend = False, ax = ax,
               color = "skyblue")

#sb aves 
sb_aves_gdf.plot(aspect = 1,
               color = "orangered",
                marker = "o",
                 alpha = 0.75,
                 markersize = 0.6,
               ax = ax)

````
